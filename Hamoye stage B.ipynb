```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
columns = {'date':'Date', 'T1':'Temp_in_kitchen','RH_1':'Humidity_in_kitchen','T2':'Temp_in_living_room',
           'RH_2':'Humidity_in_living_room','T3':'Temp_in_laundry','RH_3':'Humidity_in_laundry',
           'T4':'Temp_in_office','RH_4':'Humidity_in_office','T5':'Temp_in_bathroom','RH_5':'Humidity_in_bathroom',
           'T6':'Temp_outside_building','RH_6':'Humidity_outside_building','T7':'Temp_in_ironing_room',
           'RH_7':'Humidity_in_ironing_room','T8':'Temp_in_teenager_room','RH_8':'Humidity_in_teenager_room',
           'T9':'Temp_in_parents_room','RH_9':'Humidity_in_parents_room','rv1':'Rv1','rv2':'Rv2'}



df = pd.read_csv('energydata_complete.csv')
df.rename(columns = columns, inplace = True)
df.head()
```

```python
df.info()
```

```python
#splitting the data 
x = df['Temp_in_living_room']
y = df['Temp_outside_building']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
x = np.array(x).reshape(-1, 1)
y = np.array(y).reshape(-1, 1)
model = LinearRegression()
model.fit(x, y)
```

```python
# Predict the y values using the model
y_pred = model.predict(x)

# Calculate the R-squared value
r2 = r2_score(y, y_pred)

# Print the R-squared value rounded to two decimal places
print(f"R-squared value: {r2:.2f}")
```
